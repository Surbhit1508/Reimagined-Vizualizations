```python
import pandas as pd
import numpy as np

# First, let's examine the unique states in your dataset
print("All unique states in the dataset:")
print(df_order_stats_zip_json['Customer_State'].unique())

print("\nTotal number of states:", len(df_order_stats_zip_json['Customer_State'].unique()))

# Create a more informative state summary
state_summary = df_order_stats_zip_json.groupby('Customer_State').agg({
    'Total_Order_Amount': ['sum', 'mean'],
    'Unique_Orders': 'sum',
    'Zip_Code': 'count'
}).round(2)

# Flatten column names
state_summary.columns = ['Total Orders ($)', 'Average per State ($)', 'Number of Orders', 'Number of Zip Codes']

# Style the summary table
styled_summary = state_summary.style\
    .format({
        'Total Orders ($)': '${:,.2f}',
        'Average per State ($)': '${:,.2f}',
        'Number of Orders': '{:,.0f}',
        'Number of Zip Codes': '{:,.0f}'
    })\
    .set_properties(**{
        'background-color': '#f8f9fa',
        'color': 'black',
        'border-color': '#dee2e6',
        'padding': '8px'
    })\
    .set_table_styles([
        {'selector': 'th', 'props': [
            ('background-color', '#212529'),
            ('color', 'white'),
            ('font-weight', 'bold'),
            ('padding', '8px'),
            ('text-align', 'left')
        ]},
        {'selector': '', 'props': [
            ('border', '1px solid #dee2e6'),
            ('border-collapse', 'collapse')
        ]},
        {'selector': 'tbody tr:nth-of-type(odd)', 'props': [
            ('background-color', '#ffffff')
        ]},
        {'selector': 'tbody tr:hover', 'props': [
            ('background-color', '#f0f7ff')
        ]}
    ])\
    .highlight_max(axis=0, color='#e6fff2')\
    .highlight_min(axis=0, color='#fff0f0')

print("\nDetailed State Analysis:")
display(styled_summary)

# Add some additional insights
for state in df_order_stats_zip_json['Customer_State'].unique():
    state_data = df_order_stats_zip_json[df_order_stats_zip_json['Customer_State'] == state]
    print(f"\nState: {state}")
    print(f"Number of unique zip codes: {state_data['Zip_Code'].nunique()}")
    print(f"Average order per zip code: ${state_data['Average_Order_Amount'].mean():.2f}")
    print(f"Total revenue: ${state_data['Total_Order_Amount'].sum():.2f}")
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
from datetime import datetime, timedelta

def create_interactive_line_chart(df):
    """
    Creates an interactive, colorful line chart with multiple metrics
    """
    # Create date range for x-axis (assuming 6 months of data)
    end_date = datetime.now()
    start_date = end_date - timedelta(days=180)
    dates = pd.date_range(start=start_date, end=end_date, periods=len(df))
    
    # Create figure with secondary y-axis
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    # Add traces for different metrics
    # Revenue trend with gradient fill
    fig.add_trace(
        go.Scatter(
            x=dates,
            y=df['Total_Order_Amount'],
            name="Revenue",
            line=dict(color='#00B4D8', width=3),
            fill='tonexty',
            fillcolor='rgba(0, 180, 216, 0.2)',
            mode='lines',
            hovertemplate="<b>Date:</b> %{x|%B %d, %Y}<br>" +
                         "<b>Revenue:</b> $%{y:,.2f}<br>" +
                         "<extra></extra>"
        ),
        secondary_y=False,
    )
    
    # Average order amount trend
    fig.add_trace(
        go.Scatter(
            x=dates,
            y=df['Average_Order_Amount'],
            name="Avg Order",
            line=dict(color='#FF6B6B', width=3, dash='dot'),
            mode='lines',
            hovertemplate="<b>Date:</b> %{x|%B %d, %Y}<br>" +
                         "<b>Avg Order:</b> $%{y:,.2f}<br>" +
                         "<extra></extra>"
        ),
        secondary_y=True,
    )
    
    # Add rolling average for trend analysis
    rolling_avg = df['Total_Order_Amount'].rolling(window=7).mean()
    fig.add_trace(
        go.Scatter(
            x=dates,
            y=rolling_avg,
            name="7-Day Trend",
            line=dict(color='#4CAF50', width=2),
            mode='lines',
            hovertemplate="<b>Date:</b> %{x|%B %d, %Y}<br>" +
                         "<b>7-Day Avg:</b> $%{y:,.2f}<br>" +
                         "<extra></extra>"
        ),
        secondary_y=False,
    )
    
    # Add order volume as bubbles
    fig.add_trace(
        go.Scatter(
            x=dates,
            y=df['Unique_Orders'],
            name="Order Volume",
            mode='markers',
            marker=dict(
                size=df['Unique_Orders']/5,  # Scale bubble size
                color=df['Unique_Orders'],
                colorscale='Viridis',
                showscale=True,
                colorbar=dict(title="Orders"),
            ),
            hovertemplate="<b>Date:</b> %{x|%B %d, %Y}<br>" +
                         "<b>Orders:</b> %{y:,.0f}<br>" +
                         "<extra></extra>"
        ),
        secondary_y=False,
    )
    
    # Update layout with creative styling
    fig.update_layout(
        title={
            'text': "Interactive Revenue & Order Analysis Dashboard",
            'y':0.95,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top',
            'font': dict(size=24, color='#2C3E50')
        },
        template='plotly_white',
        hovermode='x unified',
        plot_bgcolor='rgba(255,255,255,0.9)',
        paper_bgcolor='rgba(255,255,255,0.9)',
        height=800,
        showlegend=True,
        legend=dict(
            yanchor="top",
            y=0.99,
            xanchor="left",
            x=0.01,
            bgcolor='rgba(255,255,255,0.8)',
            bordercolor='#2C3E50',
            borderwidth=1
        ),
        # Add creative shapes for visual appeal
        shapes=[
            # Add a gradient background
            dict(
                type="rect",
                xref="paper",
                yref="paper",
                x0=0,
                y0=0,
                x1=1,
                y1=1,
                fillcolor="rgba(240,242,245,0.8)",
                layer="below",
                line_width=0,
            )
        ]
    )
    
    # Update axes
    fig.update_xaxes(
        title_text="Date",
        gridcolor='rgba(128,128,128,0.1)',
        title_font=dict(size=14, color='#2C3E50'),
        tickfont=dict(size=12),
        showgrid=True,
    )
    
    fig.update_yaxes(
        title_text="Revenue ($)",
        gridcolor='rgba(128,128,128,0.1)',
        title_font=dict(size=14, color='#2C3E50'),
        tickfont=dict(size=12),
        tickformat="$,.0f",
        secondary_y=False,
        showgrid=True,
    )
    
    fig.update_yaxes(
        title_text="Average Order Amount ($)",
        gridcolor='rgba(128,128,128,0.1)',
        title_font=dict(size=14, color='#2C3E50'),
        tickfont=dict(size=12),
        tickformat="$,.0f",
        secondary_y=True,
        showgrid=False,
    )
    
    # Add range slider
    fig.update_xaxes(rangeslider_visible=True)
    
    # Add annotations for insights
    max_revenue = df['Total_Order_Amount'].max()
    max_revenue_date = dates[df['Total_Order_Amount'].argmax()]
    
    fig.add_annotation(
        x=max_revenue_date,
        y=max_revenue,
        text=f"Peak Revenue:<br>${max_revenue:,.0f}",
        showarrow=True,
        arrowhead=1,
        arrowsize=1,
        arrowwidth=2,
        arrowcolor="#FF6B6B",
        ax=0,
        ay=-40,
        font=dict(size=12, color="#2C3E50"),
        bgcolor="rgba(255,255,255,0.8)"
    )
    
    # Save the interactive HTML file
    fig.write_html("interactive_revenue_dashboard.html")
    
    return fig

# Create the visualization
try:
    fig = create_interactive_line_chart(df_order_stats_zip_json)
    fig.show()
except Exception as e:
    print(f"Error creating visualization: {e}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from datetime import datetime
import matplotlib.ticker as ticker

def create_professional_visualization(df):
    """
    Creates a clearer and more readable professional visualization with proper axis spacing
    """
    # Set the style
    plt.style.use('seaborn-whitegrid')
    sns.set_style("whitegrid", {'grid.linestyle': '--'})
    
    # Create figure and grid with larger size
    fig = plt.figure(figsize=(20, 15), dpi=300)
    gs = GridSpec(3, 2, figure=fig, height_ratios=[1, 1.2, 0.8])
    fig.patch.set_facecolor('#ffffff')
    
    # Set global font properties
    plt.rcParams['font.family'] = 'DejaVu Sans'
    plt.rcParams['font.size'] = 12
    
    # 1. Order Amount Distribution (Top Left)
    ax1 = fig.add_subplot(gs[0, 0])
    sns.kdeplot(data=df, x='Average_Order_Amount', fill=True, color='#2ecc71', ax=ax1)
    ax1.set_title('Distribution of Average Order Amount', fontsize=14, pad=20)
    ax1.set_xlabel('Average Order Amount ($)', fontsize=12)
    ax1.set_ylabel('Density', fontsize=12)
    # Format x-axis to show dollar amounts
    ax1.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
    
    # 2. State Comparison (Top Right)
    ax2 = fig.add_subplot(gs[0, 1])
    state_summary = df.groupby('Customer_State').agg({
        'Total_Order_Amount': 'sum',
        'Unique_Orders': 'sum'
    }).sort_values('Total_Order_Amount', ascending=True)
    
    colors = ['#3498db', '#2ecc71']
    bars = ax2.barh(state_summary.index, state_summary['Total_Order_Amount'], color=colors)
    ax2.set_title('Total Order Amount by State', fontsize=14, pad=20)
    ax2.set_xlabel('Total Order Amount ($)', fontsize=12)
    
    # Add value labels
    for bar in bars:
        width = bar.get_width()
        ax2.text(width * 1.02, bar.get_y() + bar.get_height()/2,
                f'${width:,.0f}',
                ha='left', va='center', fontsize=12)
    
    # Format x-axis with proper spacing
    def format_millions(x, pos):
        if x >= 1e6:
            return f'${x/1e6:.1f}M'
        elif x >= 1e3:
            return f'${x/1e3:.0f}K'
        else:
            return f'${x:.0f}'
    
    ax2.xaxis.set_major_formatter(ticker.FuncFormatter(format_millions))
    
    # Adjust spacing between ticks
    ax2.xaxis.set_major_locator(ticker.MaxNLocator(6))  # Limit number of ticks
    
    # 3. Correlation Heatmap (Middle)
    ax3 = fig.add_subplot(gs[1, :])
    numeric_cols = ['Total_Order_Amount', 'Average_Order_Amount', 'Unique_Orders',
                   'Median_Household_Income', 'Median_Age', 'Total_Population']
    correlation_matrix = df[numeric_cols].corr()
    
    sns.heatmap(correlation_matrix, annot=True, cmap='RdYlBu', center=0,
                fmt='.2f', ax=ax3, cbar_kws={'label': 'Correlation Coefficient'},
                annot_kws={'size': 10})
    ax3.set_title('Correlation Matrix of Key Metrics', fontsize=14, pad=20)
    
    ax3.set_xticklabels(ax3.get_xticklabels(), rotation=45, ha='right', fontsize=10)
    ax3.set_yticklabels(ax3.get_yticklabels(), rotation=0, fontsize=10)
    
    # 4. Summary Statistics Table (Bottom)
    ax5 = fig.add_subplot(gs[2, :])
    ax5.axis('off')
    
    summary_stats = pd.DataFrame({
        'Metric': ['Total Revenue', 'Average Order', 'Total Orders', 'Number of ZIP Codes'],
        'Value': [
            f"${df['Total_Order_Amount'].sum():,.2f}",
            f"${df['Average_Order_Amount'].mean():,.2f}",
            f"{df['Unique_Orders'].sum():,}",
            f"{len(df):,}"
        ]
    })
    
    table = ax5.table(cellText=summary_stats.values,
                     colLabels=summary_stats.columns,
                     loc='center',
                     cellLoc='center',
                     colColours=['#f0f0f0']*2,
                     cellColours=[['#f8f9fa']*2]*len(summary_stats))
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    table.scale(1.5, 2)
    
    # Add timestamp watermark
    fig.text(0.99, 0.01, f'Created: {datetime.now().strftime("%Y-%m-%d %H:%M")}',
             fontsize=8, color='gray', ha='right', alpha=0.5)
    
    # Adjust layout with more space
    plt.tight_layout(pad=3.0)
    
    # Save visualization
    plt.savefig('order_analysis_visualization.png', dpi=300, bbox_inches='tight')
    
    return fig

# Create visualization with error handling
try:
    viz_figure = create_professional_visualization(df_order_stats_zip_json)
    plt.show()
except Exception as e:
    print(f"Error creating visualization: {e}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from datetime import datetime

def create_executive_visualization(df):
    """
    Creates business-friendly visualization focused on key insights
    """
    plt.style.use('seaborn')
    fig = plt.figure(figsize=(20, 15), dpi=300)
    gs = GridSpec(2, 2, figure=fig)
    fig.patch.set_facecolor('#ffffff')
    
    # Set global font properties
    plt.rcParams['font.family'] = 'DejaVu Sans'
    plt.rcParams['font.size'] = 12
    
    # 1. State Revenue Comparison (Top Left)
    ax1 = fig.add_subplot(gs[0, 0])
    state_data = df.groupby('Customer_State').agg({
        'Total_Order_Amount': 'sum',
        'Unique_Orders': 'sum'
    }).sort_values('Total_Order_Amount', ascending=False)  # Corrected parameter
    
    colors = ['#2ecc71', '#3498db']
    bars = ax1.bar(state_data.index, state_data['Total_Order_Amount'], color=colors)
    ax1.set_title('Revenue by State', fontsize=14, pad=20)
    ax1.set_ylabel('Total Revenue ($)', fontsize=12)
    
    # Add value labels with better formatting
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, height * 1.01,
                f'${height:,.0f}',
                ha='center', va='bottom', fontsize=12)
    
    # Format y-axis with dollar signs and commas
    ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))
    
    # 2. Customer Base Distribution (Top Right)
    ax2 = fig.add_subplot(gs[0, 1])
    order_counts = df.groupby('Customer_State')['Unique_Orders'].sum()
    total_orders = order_counts.sum()
    percentages = order_counts / total_orders * 100
    
    wedges, texts, autotexts = ax2.pie(percentages, 
                                      labels=['Texas', 'Arkansas'],
                                      autopct='%1.1f%%',
                                      colors=colors,
                                      explode=(0.05, 0))
    ax2.set_title('Distribution of Orders by State', fontsize=14, pad=20)
    
    # Make percentage labels larger
    plt.setp(autotexts, size=12, weight="bold")
    plt.setp(texts, size=12)
    
    # 3. Key Business Metrics (Bottom)
    ax3 = fig.add_subplot(gs[1, :])
    ax3.axis('off')
    
    # Create business-focused metrics with insights
    metrics = pd.DataFrame({
        'Key Performance Indicators': [
            'Total Revenue',
            'Average Order Value',
            'Total Number of Orders',
            'Geographic Coverage',
            'Market Performance',
            'Customer Demographics'
        ],
        'Values': [
            f"${df['Total_Order_Amount'].sum():,.2f}",
            f"${df['Average_Order_Amount'].mean():,.2f}",
            f"{df['Unique_Orders'].sum():,}",
            f"{len(df):,} ZIP Codes",
            f"TX: {percentages['TX']:,.1f}% | AR: {percentages['AR']:,.1f}%",
            f"Avg Household Income: ${df['Median_Household_Income'].mean():,.2f}"
        ],
        'Business Insights': [
            "Combined revenue across both markets",
            "Average customer transaction value",
            "Total successful transactions completed",
            "Current market penetration by ZIP code",
            f"Texas leads with {percentages['TX']:,.1f}% of total orders",
            "Target market purchasing power indicator"
        ]
    })
    
    # Create and format table
    table = ax3.table(cellText=metrics.values,
                     colLabels=metrics.columns,
                     loc='center',
                     cellLoc='left',
                     colColours=['#e8f4f9']*3,
                     cellColours=[['#ffffff']*3]*len(metrics))
    
    # Format table
    table.auto_set_font_size(False)
    table.set_fontsize(11)
    table.scale(1.5, 2)
    
    # Adjust column widths
    col_widths = [0.3, 0.3, 0.4]
    for idx, width in enumerate(col_widths):
        table.auto_set_column_width([idx])
    
    # Add title
    fig.suptitle('Business Performance Summary', fontsize=16, y=0.95)
    
    # Add timestamp
    fig.text(0.99, 0.01, f'Report Generated: {datetime.now().strftime("%Y-%m-%d")}',
             fontsize=8, color='gray', ha='right')
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    # Save visualization
    plt.savefig('business_performance_summary.png', dpi=300, bbox_inches='tight')
    
    return fig

# Create visualization
viz_figure = create_executive_visualization(df_order_stats_zip_json)
plt.show()
```
